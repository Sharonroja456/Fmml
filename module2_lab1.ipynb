```python
import requests
url = "https://raw.githubusercontent.com/SP9144/FMML_Data/main/Automobile_data.csv"
response = requests.get(url)
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
from io import StringIO
import matplotlib.pyplot as plt
```

```python
data = StringIO(response.text)
data = pd.read_csv(data)
print("Number of data points =", len(data))
data.head()
```

```python
data.info()
```

```python
# Removing rows which have missing values

data.replace("?", np.nan, inplace = True)
data = data.dropna()
print(len(data))
```

```python
# Dividing the dataset into features (X) and the target variable (y)

X = data.drop("price", axis = 1)
y = pd.to_numeric(data["price"])
```

```python
plt.figure(figsize = (24, 7))
plt.scatter(X["make"], y,color="red")
plt.xlabel('Make')
plt.ylabel('Price')
plt.title('Car maker vs Price - Scatter Plot')
plt.show()
```

# Swarm plot

```python
sns.swarmplot(x=X["make"],y=y,color="red").set_title('Car maker vs Price Swarm Plot')
```

# join plot

```python
sns.jointplot(x=pd.to_numeric(X["horsepower"]), y = y, kind="reg", color = 'red')
```

#Box plot
A Box plot is a way to visualize the distribution of the data by using a box and some vertical lines. It is known as the whisker plot. The data can be distributed between five key ranges, which are as follows:

Minimum: Q1-1.5*IQR

1st quartile (Q1): 25th percentile

Median:50th percentile

3rd quartile(Q3):75th percentile

Maximum: Q3+1.5*IQR
